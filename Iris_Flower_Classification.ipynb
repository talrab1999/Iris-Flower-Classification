```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df['Species'].unique()
```

```python
iris_type_map = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}
df['Species'] = df['Species'].map(iris_type_map)
df.head()
```

```python
df['Species'].unique()
```

```python
for i in range(len(df.columns) - 1):
  for j in range(i + 1, len(df.columns) - 1):
    x_label = df.columns[i]
    y_label = df.columns[j]
    sns.scatterplot(x = x_label, y = y_label, data = df, hue = 'Species', palette= ['red', 'blue', 'green'])
    plt.show()
```

```python
sns.pairplot(df, hue='Species', palette= ['red', 'blue', 'green'])
```

```python
df.columns
```

```python
train = df[df.columns[:-1]]
true_vals = df[df.columns[-1:]]
print(train)
print(true_vals)
```

```python
#dropping ID col because it's cheating the dataset
df = df.drop(columns= ['Id'])
df.head()
```

#Clustering K-means

```python
from sklearn.cluster import KMeans
```

```python
x = 'SepalLengthCm'
y = 'PetalWidthCm'
X = train[[x,y]].values
```

```python
kmeans = KMeans(n_clusters= 3)
kmeans = kmeans.fit(X)
```

```python
clusters = kmeans.labels_
```

```python
clusters
```

```python
true_vals.values.reshape(1,-1)
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns = [x, y, 'Species'])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(x = x, y = y, hue = 'Species', data = cluster_df, palette= ['red', 'blue', 'green'], ax = ax1)
ax1.set_title('Prediction')
sns.scatterplot(x = x, y = y, hue = 'Species', data = df, palette= ['red', 'blue', 'green'], ax = ax2)
ax2.set_title('True Values')
plt.show()
```

#Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
clusters = gmm.fit_predict(X)
clusters
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns = df.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(x = x, y = y, hue = 'Species', data = cluster_df, palette= ['red', 'blue', 'green'], ax = ax1)
ax1.set_title('Prediction')
sns.scatterplot(x = x, y = y, hue = 'Species', data = df, palette= ['red', 'blue', 'green'], ax = ax2)
ax2.set_title('True Values')
plt.show()
```

that's some crazy accuracy

```python
true_vals.values.reshape(1, -1)
true_vals = true_vals.values.ravel()
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_map = {0: 3, 1: 1, 2: 2}
mapped_clusters = np.vectorize(acc_map.get)(clusters)
```

```python
accuracy = accuracy_score(true_vals, mapped_clusters)
print(f"Accuracy: {accuracy:.2f}")
```

#Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X_pca)
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns = df.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(x = x, y = y, hue = 'Species', data = cluster_df, palette= ['red', 'blue', 'green'], ax = ax1)
ax1.set_title('Prediction')
sns.scatterplot(x = x, y = y, hue = 'Species', data = df, palette= ['red', 'blue', 'green'], ax = ax2)
ax2.set_title('True Values')
plt.show()
```
